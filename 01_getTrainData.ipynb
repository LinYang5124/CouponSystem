```python
sql_1 = 'select * from campaign_data'
cursor.execute(sql_1)
results = cursor.fetchall()
campaign = pd.DataFrame(data = results)
```

```python
campaign_total = pd.DataFrame(columns = ['campaign_quantity',
                                          'campaign_selling_price',
                                          'campaign_other_discount', 
                                          'campaign_coupon_discount'])
```

```python
#campaign_total
```

```python
for i in range(len(campaign)):
#for i in range(4):
    camp = campaign.iloc[i]
    sql_2 = 'select  sum(quantity) as campaign_quantity, \
                sum(selling_price) as campaign_selling_price, \
                sum(other_discount) as campaign_other_discount, \
                sum(coupon_discount) as campaign_coupon_discount \
             from    customer_transaction_data  \
             where date>=\'{0}\' and date<=\'{1}\''.format(camp['start_date'], camp['end_date'])
    #print(sql_2)
    cursor.execute(sql_2)
    results = cursor.fetchall()
    campaign_total = pd.concat([campaign_total,pd.DataFrame(data = results)],axis=0)
    print(str(i)+' is OK!')
```

```python
campaign_total = campaign_total.set_index([np.arange(0,len(campaign))])
campaign = pd.concat([campaign, campaign_total], axis=1)
campaign
```

```python
trains = pd.merge(left=trains, right=campaign, on='campaign_id', how='left')
```

```python
sql_3 = 'select customer_id, \
            sum(quantity) as customer_quantity, \
            sum(selling_price) as customer_selling_price, \
            sum(other_discount) as customer_other_discount, \
            sum(coupon_discount) as customer_coupon_discount \
         from customer_transaction_data \
         group by customer_id'
cursor.execute(sql_3)
results = cursor.fetchall()
customer_total = pd.DataFrame(data = results)
customer_total
```
